```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv(r"/content/SEE_Class11_Class12_Statistics_2009_2016.csv")
dataframe
```

```python
dataframe.shape
```

```python
dataframe.info()
```

#lets compare the appared boys and girls in CLASS-11


```python
dataframe["C11_Appeared_Boys"]
sns.barplot(x="Year",y="C11_Appeared_Boys",data=dataframe)
plt.title("Number of Boys in Class-11")
```

Above we can see the highest of boys appeared in class 11 was year 2015 and the least one was the year 2014.

```python
dataframe["C11_Appeared_Girls"]
sns.barplot(x="Year",y="C11_Appeared_Girls",data=dataframe)
plt.title("Number of Girls in Class-11")
```

Above we can see the highest of girls appeared in class 11 was year 2015 and the least one was the year 2014.

```python
sns.barplot(x="Year",y="C11_Appeared_Boys",data=dataframe, label="BOYS",color="gray" )
sns.barplot(x="Year",y="C11_Appeared_Girls",data=dataframe,label="GIRLS",color="red")
plt.title("Number of Boys and Girls in Class-11")
```

This above is the comparision of girls and boys in same graph line.

```python

x = np.arange(len(dataframe["Year"]))
width = 0.4

plt.bar(x - width/2, dataframe["C11_Appeared_Boys"], width, label="Boys", color="gray")
plt.bar(x + width/2, dataframe["C11_Appeared_Girls"], width, label="Girls", color="red")

plt.xticks(x, dataframe["Year"])
plt.title("Number of Boys and Girls in Class 11")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

The highest number of students appeared is 2015 for both boys and girls and least is 2014 for both boys and girls in class 11.

```python

x = np.arange(len(dataframe["Year"]))
width = 0.4

plt.bar(x - width/2, dataframe["C11_Passed_Boys"], width, label="Boys", color="green")
plt.bar(x + width/2, dataframe["C11_Passed_Girls"], width, label="Girls", color="yellow")

plt.xticks(x, dataframe["Year"])
plt.title("Number of Passed Boys and Girls in Class 11")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

Highest number of passed students of class 11 for both boys and girls is year 2015 and least is  year 2009 for both boys and girls.
