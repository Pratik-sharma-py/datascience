```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/archive (1).zip')
df.head()
```

```python
df.head(10)
```

#we are going to apply math in this data from mean to correlation all but at first we are taking two columns to find correlation is attendance and exam score

```python
df_attendance = df['Attendance'].tolist()
df_score = df['Exam_Score'].tolist()


attendance_num = len(df_attendance)
print(attendance_num)

score_num = len(df_score)
print(score_num)
```

From the above output we know that Number of items(n) is 6607

```python
#lets find the range of attendance

attendance_range = max(df_attendance)-min(df_attendance)
print("Range of attendance :",attendance_range)

#lets find the range of attendance

score_range = max(df_score)-min(df_score)
print("Range of Score :", score_range)
```

```python
#lets find the mean of attendance

attendance_mean = sum(df_attendance)/attendance_num
print("Mean of the attendance :",attendance_mean)

#lets find the mean of score

score_mean = sum(df_score)/score_num
print("Mean of the score :",score_mean)
```

```python
#lets find the varience of the attendance

attendance_varience = 0
for i in df_attendance:
  attendance_varience += (i-attendance_mean)**2
attendance_varience = attendance_varience/attendance_num
print("Varience of the attendance :",attendance_varience)

#lets find the varience of the score

score_varience = 0
for i in df_score:
  score_varience += (i-score_mean)**2
score_varience = score_varience/score_num
print("Varience of the Score :", score_varience)
```

```python
#lets find the standard deviation of attendance

attendance_std = attendance_varience**0.5
print("Standard Deviation of attendance :",attendance_std)

#lets find the standard deviation of score

score_std = score_varience**0.5
print("Standard Deviation of Score :", score_std)
```

```python
#lets find the median of the attendance

df_attendance.sort()
if attendance_num%2 == 0:
  median = (df_attendance[attendance_num//2] + df_attendance[attendance_num//2-1])
  print("Median of the attendance :",median)
else:
  median = (df_attendance[attendance_num//2])
  print("Median of the attendance :", median)

#lets find the median of the score

df_score.sort()
if score_num%2 == 0:
  median = (df_score[score_num//2] + df_score[score_num//2-1])
  print("Median of the score :", median)
else:
  median = (df_score[score_num//2])
  print("Median of the score :", median)

```

```python
#lets find the lower half and upper half of attendance

lower_half1 = []
upper_half1 = []
for i in df_attendance:
  if i < attendance_mean:
    lower_half1.append(i)
  else:
    upper_half1.append(i)


#lets find the lower half and upper half of score

lower_half2 = []
upper_half2 = []
for i in df_score:
  if i < score_mean:
    lower_half2.append(i)
  else:
    upper_half2.append(i)
```

```python
#lets find the Q1 and Q3 of the attendance

Q1_attendance = np.median(lower_half1)
print("Q1 of the attendance :",Q1_attendance)

Q3_attendance = np.median(upper_half1)
print("Q3 of the attendance :", Q3_attendance)


#lets find the Q1 and Q3 of the score

Q1_score = np.median(lower_half2)
print("Q1 of the score :", Q1_score)

Q3_score = np.median(upper_half2)
print("Q3 of the score :", Q3_score)
```

```python
#lets find the inter quartile range (IQR) of attendance

IQR_attendance = Q3_attendance - Q1_attendance
print("IQR of the attendance :", IQR_attendance)

#lets find the inter quartile range (IQR) of score

IQR_score = Q3_score - Q1_score
print("IQR of the score :", IQR_score)
```

```python
#lets find the covareience of the attendance and score

covriance = 0
for i in range(attendance_num):
  covriance += ((df_attendance[i]-attendance_mean)*(df_score[i]-score_mean))
covriance = covriance/attendance_num

print("Covriance of attendance and score :", covriance)
```

```python
#lets find the correlation of attendance and score

correlation = covriance/(attendance_std*score_std)
print("Correlation of attendance and score is :", correlation)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_attendance, df_score, color='blue')
plt.xlabel("Attendance")
plt.ylabel("Score")
plt.title(f"Correlation: {correlation:.2f}")
plt.show()
```
