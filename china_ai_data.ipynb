```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

AI_Startups_Count = 	Number of startups contributing to innovation and development

This dataset contains yearly data (2010–2026) for China, including key indicators of research, investment, innovation, and growth for analysis and modeling.

```python
dataset= pd.read_csv(r"/content/archive (12).zip")
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
clean_dataset = dataset.copy()
```

```python
clean_dataset.columns = clean_dataset.columns.str.replace(r' \(%\)', '', regex=True).str.replace(' ', '_').str.lower().str.strip()
clean_dataset
```

```python
clean_dataset.info()
```

**AI_Publication_Share (%) = Percentage share of global research publications, reflecting research contribution**

```python

# 2. Create the plot
plt.figure(figsize=(14, 10)) # Optional: Sets figure size

# Fill the area below the line with a semi-transparent color for a 'blurry' effect
plt.fill_between(clean_dataset['year'], clean_dataset['ai_publication_share'], color='skyblue', alpha=0.4)

# Plot the line on top of the filled area
plt.plot(clean_dataset['year'], clean_dataset['ai_publication_share'] , marker='o', linestyle='-', color='darkblue', linewidth=2, label='AI Publication Share')

# 3. Add Details (Labels, Title, Grid)
plt.title('AI Publication Share Trends Over the Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('AI Publication Share', fontsize=12)
plt.grid(True)
plt.legend()

# 4. Display Plot
plt.show()
```

AI_Investment_Billion_USD = Total annual investment in technology measured in billion USD

Government_AI_Spending_Billion_USD = Government expenditure on research and development activities

Private_AI_Investment_Billion_USD = Investment made by private sector organizations

```python
plt.figure(figsize=(14, 10))
sns.barplot(x="year",y="ai_investment_billion_usd",data=clean_dataset, label="Total",color="green" )
sns.barplot(x="year",y="government_ai_spending_billion_usd",data=clean_dataset,label="Government",color="blue")
sns.barplot(x="year",y="private_ai_investment_billion_usd",data=clean_dataset,label="Private",color="yellow")


plt.title("Government vs. Private AI Investment Over the Years (Billion USD)'")
```

**R&D_Expenditure (% of GDP) = National spending on research and development as a percentage of GDP**

```python
# 2. Create the plot
plt.figure(figsize=(14, 10)) # Optional: Sets figure size

# Fill the area below the line with a semi-transparent color for a 'blurry' effect
plt.fill_between(clean_dataset['year'], clean_dataset['r&d_expenditure_(%_of_gdp)'], color='skyblue', alpha=0.4)

# Plot the line on top of the filled area
plt.plot(clean_dataset['year'], clean_dataset['r&d_expenditure_(%_of_gdp)'] , marker='o', linestyle='-', color='darkblue', linewidth=2, label='r&d_expenditure_(%_of_gdp)')

# 3. Add Details (Labels, Title, Grid)
plt.title('National spending on research and development as a percentage of GDP', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('r&d_expenditure_(%_of_gdp) ', fontsize=12)
plt.grid(True)
plt.legend()

# 4. Display Plot
plt.show()
```

**AI_Growth_Rate (%) = Annual growth rate calculated based on investment changes (used as target variable)**

```python
# 2. Create the plot
plt.figure(figsize=(14, 10)) # Optional: Sets figure size

# Fill the area below the line with a semi-transparent color for a 'blurry' effect
plt.fill_between(clean_dataset['year'], clean_dataset['ai_growth_rate'], color='skyblue', alpha=0.4)

# Plot the line on top of the filled area
plt.plot(clean_dataset['year'], clean_dataset['ai_growth_rate'] , marker='o', linestyle='-', color='darkblue', linewidth=2, label='AI Growth Rate')

# 3. Add Details (Labels, Title, Grid)
plt.title('AI Growth Rate of China 2010-2026', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('AI Growth Rate in %', fontsize=12)
plt.grid(True)
plt.legend()

# 4. Display Plot
plt.show()
```
