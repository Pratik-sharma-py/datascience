#Types of Operators

An operator is a symbol that performs a certain operation between operands.



Arithmetic — (+ - * / // % **) — math on numbers.

Comparison — (== != > < >= <=) — compare values, always return True or False.

Assignment — (= += -= *= etc. )— assign or update variables in place.

Logical — (and or not )— combine or negate boolean conditions.

Bitwise — (& | ^ ~ << >> )— operate directly on binary bits of integers.

Identity — (is / is not) — check whether two variables point to the exact same object in memory (not just equal values).

Membership — (in / not in) — check whether a value exists inside a list, string, tuple, dict, etc.

Walrus := — introduced in Python 3.8, assigns a value and returns it in a single expression, useful inside while loops and comprehensions.

#Recursion

**When a function calls itself repeatedly.**

````python
#print n to 1 backwards

def show(n):
  if(n==0): #Base case
    return
  print(n)
  show(n-1)

show(5)
````

````python
#return n!

def fact(n):
  if(n == 0 or n == 1): #base case > stops the recursion
    return 1
  else:
    return n * fact(n-1)

fact(10)
````

````python
# Write a recursive function to calculate the sum of first n natural numbers.

def sum_natural(n):
  if(n==0):
    return 0
  else:
    return n + sum_natural(n-1)

sum_natural(15)

````

#Super Method

**super() method is used to access methods of the parent class.**

````python
class Car:
  def __init__(self,type):
    self.type = type


    @staticmethod
    def start():
      print("car started..")

    @staticmethod
    def stop():
      print("car stopped.")

class ToyotaCar(Car):
  def __init__(self,name,type):
    super().__init__(type)  # super calls the methods from the parent class.
    self.name = name

car1 = ToyotaCar("prius","electric")
print(car1.type)
````

**order1>order2> if price of order1>price of order2**

````python
class Order():
  def __init__(self,item,price):
    self.item = item
    self.price = price

  def __gt__(self,odr2):
    return self.price > odr2.price

odr1 = Order("egg","500")
odr2 = Order("museli","575")

print(odr1 > odr2)
````
