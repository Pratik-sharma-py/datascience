```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_tourist = pd.read_csv("/content/tourist_arrival_data.csv")
df_tourist.head(15)
```

```python
df_tourist.describe()
```

```python
df_tourist.info()
```

```python
#Find the total number of people arrived each year

total_tourist = df_tourist.groupby('Year')['Arrivals'].sum()
total_tourist
```

```python
#Average toursit arrived in a year

total_tourist.mean()
```

```python
#Median of toursit arrived

total_tourist.median()
```

Mean > Median so we can say the data is slightly right skewed because of the outlier of year 2026 because it contains only one month of data.

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(total_tourist.index, total_tourist.values, marker='o', linestyle='-')
ax.set_xlabel('Year')
ax.set_ylabel('Total Arrivals')
ax.set_title('Total Tourist Arrivals by Year (Line Plot)')
ax.set_xticks(total_tourist.index)
ax.set_xticklabels(total_tourist.index, rotation=45)
ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Find the top 10 years of arrival

total_top10_tourist = total_tourist.sort_values(ascending=False).head(10)
total_top10_tourist
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

x_positions = range(len(total_top10_tourist))
ax.bar(x_positions, total_top10_tourist.values, width=0.6, edgecolor="white", linewidth=0.7, color='green')
ax.set_xlabel('Year')
ax.set_ylabel('Total Arrivals')
ax.set_title('Total Highest Tourist Arrivals by Year')
ax.set_xticks(x_positions)
ax.set_xticklabels(total_top10_tourist.index, rotation=45, ha='right')  # ha='right' aligns rotated labels properly

for i, value in enumerate(total_top10_tourist.values):
    ax.text(i, value, f'{value:,}', ha='center', va='bottom', fontsize=10)

plt.subplots_adjust(bottom=0.18)
plt.show()
```

```python
#Find the total number of people arrived in each month

total_tourist_month = df_tourist.groupby(['Month','Year'])['Arrivals'].sum()
total_tourist_month
```

```python
#Find the month where most tourist arrived

total_tourist_month_top10 = total_tourist_month.sort_values(ascending=False).head(10)
total_tourist_month_top10
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(total_tourist_month_top10.index.map(str), total_tourist_month_top10.values, width=0.8, edgecolor="white", linewidth=0.7,color='green')
ax.set_xlabel('Month / Year')
ax.set_ylabel('Total Arrivals')
ax.set_title('Total TOP10 Tourist Arrivals by Month in Year')
ax.set_xticks(range(len(total_tourist_month_top10.index)))
# Convert MultiIndex tuples to strings for x-axis labels
x_labels = [f'{month} - {year}' for month, year in total_tourist_month_top10.index]
ax.set_xticklabels(x_labels, rotation=45)

# Add value labels on top of each bar
for i, value in enumerate(total_tourist_month_top10.values):
    ax.text(i, value, f'{value:,}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Many tourists visit Nepal in October, November, and March due to the prime autumn/spring seasons, which offer the best weather for trekking, adventure sports, and mountain views. These months provide clear skies and moderate temperatures, making them ideal for mountaineering, cultural tours, and witnessing peak festival seasons, notably in October.


#Key Factors for November 2018 Success:

Peak Season Timing: November is peak tourism season in Nepal, offering excellent trekking weather.

Milestone Year: By November 2018, cumulative arrivals reached 1,001,930, marking the first time in history that tourism in Nepal crossed the 1-million mark.

Significant Inflow: Over 116,553 international visitors entered Nepal in November 2018, contributing heavily to a 17% overall increase in arrivals for that year.

Top Markets: High numbers of visitors originated from India (16,167), China (12,944), and the US, driving the November surge.

Promotional Efforts: The success resulted from joint marketing initiatives by the Nepal Tourism Board and private sector stakeholders aimed at building confidence for Visit Nepal Year 2020.

```python
#Calculate year to year growth

year_to_year_growth = total_tourist.pct_change() * 100
year_to_year_growth
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(year_to_year_growth.index, year_to_year_growth.values, marker='o', linestyle='-')
ax.set_xlabel('Year')
ax.set_ylabel('Arrivals')
ax.set_title('Year to year growth  (Line Plot)')
ax.set_xticks(year_to_year_growth.index)
ax.set_xticklabels(year_to_year_growth.index, rotation=45)
ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
monthly_arrivals_df = total_tourist_month.unstack(level='Month')

# Define a list of months in chronological order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Reindex the columns to ensure months are in chronological order
monthly_arrivals_df = monthly_arrivals_df.reindex(columns=month_order)

monthly_arrivals_df.head()
```

```python
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.heatmap(monthly_arrivals_df, cmap='viridis', annot=True, fmt='.0f', linewidths=.5, linecolor='black')
plt.title('Heatmap of Monthly Tourist Arrivals (Years vs. Months)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.yticks(rotation=0)
plt.show()
```
