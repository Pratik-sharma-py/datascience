```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('/content/archive.zip')
df.head(10)
```

```python
Quantity_list = df['quantity_sold'].tolist()
rating_list = df['rating'].tolist()

quan_num = len(Quantity_list)
print(quan_num)
rating_num = len(rating_list)
print(rating_num)



```

Now we are going to find the mean of Quantity_list and Rating_list

```python
#mean of quantity_list
mean_quantity = sum(Quantity_list)/quan_num
print("mean of quantity sold :",mean_quantity)

#mean of rating_list

mean_rating = sum(rating_list)/rating_num
print("mean of rating :", mean_rating)
```

Now,  we are going to find out the covareience Cov(quantity_list,rating_list)

```python
covariance = sum((Quantity_list[i] - mean_quantity) * (rating_list[i] - mean_rating) for i in range(quan_num)) / (quan_num - 1)
print("covariance:", covariance)

```

Calculating standard deviation

```python
#standard deviation of Quantity_list
std_Quantity_list_sum = 0
for i in range(quan_num):
    std_Quantity_list_sum += (Quantity_list[i] - mean_quantity)**2
std_Quantity_list = (std_Quantity_list_sum / (quan_num - 1))**0.5
print("Standard Deviation of Quantity sold :", std_Quantity_list)

#Standard Deviation of rating_list
std_rating_list_sum = 0
for i in range(rating_num):
  std_rating_list_sum += (rating_list[i] - mean_rating)**2
std_rating_list = (std_rating_list_sum / (rating_num -1))**0.5
print("Standard Deviation of Rating : ", std_rating_list)

```

Calculating Correlation between them

```python
correlation = covariance / (std_Quantity_list * std_rating_list)
print("Correlation between Quantity sold and rating:", correlation)
```
