#Lets Start cleaning the data

```python
# Make a copy so the original raw data stays unchanged.

clean_df = df.copy()
```

```python
df.dtypes
```

```python
main_columns = ["show_id","type","title","director","cast","country","date_added","release_year","rating","duration","listed_in","description"]
for col in main_columns:
  clean_df[col] = clean_df[col].astype("string").str.strip()

#clean_df
```

```python
clean_df["country"] = clean_df["country"].str.lower()
clean_df["country"]
```

```python
clean_df.info()
```

```python
clean_df.dtypes
```

```python
# Fix 1: Convert date to real datetime
clean_df['date_added'] = pd.to_datetime(
    clean_df['date_added'], errors='coerce')

clean_df['year_added'] = clean_df['date_added'].dt.year
clean_df['month_added'] = clean_df['date_added'].dt.month

clean_df.head()
```

```python
# Fix 2: Fill missing text fields
clean_df['director'] = clean_df['director'].fillna('Unknown')
clean_df['cast'] = clean_df['cast'].fillna('Unknown')
clean_df['country'] = clean_df['country'].fillna('Unknown')
clean_df
```

```python
# Fix 3: Remove dirty values from rating
bad_ratings = ['74 min', '84 min', '66 min']
clean_df.loc[clean_df['rating'].isin(bad_ratings), 'rating'] = None

clean_df['rating'] = clean_df['rating'].fillna('Unknown')
clean_df
```

```python
# Fix 4: Separate numeric duration for movies
# 90 min = 90 (number)     2 Seasons  = 2 (number)
movie_mask = clean_df['type'] == 'Movie'
show_mask  = clean_df['type'] == 'TV Show'

clean_df.loc[movie_mask, 'duration_mins'] = (
    clean_df.loc[movie_mask, 'duration'].str.extract(r'(\d+)')[0].astype(float)
)
clean_df.loc[show_mask, 'duration_seasons'] = (
    clean_df.loc[show_mask, 'duration'].str.extract(r'(\d+)')[0].astype(float)
)

```
